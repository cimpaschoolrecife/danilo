## 2. Creating Arrays
Arrays can be created from Python lists or using NumPy helper functions.

**Example:**

```python
a = np.array([1, 2, 3])
b = np.array([[1, 2], [3, 4]])
```

````python
a = np.array([1, 2, 3])
b = np.array([[1, 2], [3, 4]])
print(a)
print(b)
````

### Common Array Creation Functions

**np.zeros:** Creates a $nxm$ array of zero values:

Example:
```python
zeros = np.zeros((2, 3))
```
**np.ones:** Creates a $nxm$ array of 1's:

Example:
```python
ones = np.ones((3, 3))
```
**np.eye:** Creates a $nxn$ identity matrix:

Example:
```python
identity = np.eye(3)
```

````python
zeros = np.zeros((2, 3))
ones = np.ones((3, 3))
identity = np.eye(3)
print(zeros)
print(ones)
print(identity)
````

## 4. Basic Element-wise Operations

Given two $nxm$ arrays, $x$ and $y$, you can perform element-wise sum, subtraction and multiplication by computing:

```python
x + y
x - y
x * y
```

````python
x = np.array([1, 2, 3])
y = np.array([4, 5, 6])
print(x + y)
print(x - y)
print(x * y)
````

## 6. Matrix Multiplication

````python
A = np.array([[1, 2], [3, 4]])
B = np.array([[5, 6], [7, 8]])
print(A @ B)
print(np.dot(A, B))
````

## 10. Linear Algebra Concepts
NumPy provides a submodule `np.linalg` for linear algebra operations.

### Transpose

````python
print(A.T)
````

### Determinant

````python
print(np.linalg.det(A))
````

### Inverse Matrix

````python
print(np.linalg.inv(A))
````

### Eigenvalues and Eigenvectors

````python
eigenvalues, eigenvectors = np.linalg.eig(A)
print(eigenvalues)
print(eigenvectors)
````
